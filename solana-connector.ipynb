```python
def getKeys(block):
    # Get all the unique account keys used in one block
    keys = []
    for T in block['result']['transactions']:
        for key in T['transaction']['message']['accountKeys']:
            if key not in keys:
                keys.append(key)
    return keys
    
```

```python
len (getKeys(block))
```

```python
def allKeys(block):
    # Get all the unique account keys used in one block
    keys = []
    for T in block['result']['transactions']:
        for key in T['transaction']['message']['accountKeys']:
            keys.append(key)
    return keys

def most_common(lst):
    return max(set(lst), key=lst.count)
```

```python
all_keys = allKeys(block)

most_common(all_keys)
```
